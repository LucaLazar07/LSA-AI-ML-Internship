**Task 3**

```python
# HP from cars.xslx
numeric_list = [49, 55, 92, 70, 81, 62, 53, 78, 103, 130, 115, 118]

# the selected range is 80 - 150
def check_range(n):
  if 80 <= n <= 150:
    return 1
  return 0
total_values = 0
invalid_values = 0
for number in numeric_list:
  total_values += 1
  if check_range(number):
    print(f"Number {number} is within the acceptable range")
  else:
    invalid_values += 1
    print(f"Number {number} is not within the acceptable range")
print(f"Total values checked: {total_values}\nInvalid values: {invalid_values}")
```

**Task 4**

```python
import pandas as pd

dataset = pd.read_excel("cars.xlsx")

# Describe the data
dataset.describe()
```

```python
# Print first five rows
dataset.head(5)
```

```python
import matplotlib.pyplot as plt

# Statistical visualisation of the data
for data in dataset.columns.tolist():
  plt.hist(dataset[data].dropna(), bins = 15)
  plt.xlabel("Data Values")
  plt.title(f"{data} visualisation")
  plt.ylabel("Frequnecy")
  plt.show()
```

```python
plt.scatter(dataset["HP"], dataset["VOL"], color="blue")

plt.xlabel("Horse Power")
plt.ylabel("Volume")
plt.title("Volume as a function of Horse Power")
plt.show()
```

```python
plt.scatter(dataset["SP"], dataset["MPG"], color="red")

plt.xlabel("Specific Power")
plt.ylabel("Miles Per Gallon")
plt.title("MPG as a function of SP")
plt.show()
```

```python
plt.boxplot(dataset["WT"])
plt.show()
```
